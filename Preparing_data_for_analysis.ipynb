**Reading a new dataset Calorie_df and preparing it for analysis**

```python
import pandas as pd
url = "https://github.com/futureCodersSE/working-with-data/blob/main/Data%20sets/Data%20Cleaning%20Data%20Sets/dirtydata.csv?raw=true"
calorie_df=pd.read_csv(url)

calorie_df
```

In the below example, I have used .loc() to locate a record with incorrect value and have assigned a correct value to it.

```python
#to replace incorrect value for Duration with the right one

calorie_df1=calorie_df
calorie_df1.loc[calorie_df1['Duration'] ==450, 'Duration'] = 45
calorie_df1
```

Below, I have used df.info() to check which columns contain null values.

```python
calorie_df.info()
```

Below I have replaced null values by for calories column by average value. I have used groupby('Duration') as I wanted to fill null values with average calories for the perticular duration.

```python
#Fill null values by average value grouped by duration

calorie_df2=calorie_df1
calorie_df2['Calories']=calorie_df2.groupby('Duration')['Calories'].transform(lambda x:x.fillna(x.mean()))
calorie_df2['Calories'] = calorie_df2['Calories'].apply(lambda x: round(x, 1))
calorie_df2
```

In the below example,
1. I have used drop_duplicates to remove any duplicate records
2. I have used .fillna to replace null value with a correct value
3. I have used to_datetime function to convert date values to correct datatype.

```python
calorie_df3 = calorie_df2

calorie_df3=calorie_df3.drop_duplicates() #to remove duplicate record

calorie_df4=calorie_df3
calorie_df4['Date'].fillna("2020-12-22", inplace=True) #to replace null value with correct value

calorie_df4['Date']=pd.to_datetime(calorie_df4['Date']) #to convert values from date column to correct datatype
calorie_df4
```

**Normalising total vaccinations**

I wanted to find the minimum total vaccinations for the United Kingdom. So, first I created a new dataframe with only two columns, country and minimum total vaccinations for each country. I used .groupby('country').min() to find out minimum total vaccinations for each country. Then, I used df.loc with subset 'united kingdom' to find out min vaccinations for United Kingdom. I used int() function to round this min vaccinations for UK value.
Then, I wrote a function to normalise total_vaccinations column so that all values less than the UK's min are 0 and all values greater than or equal to the UK's min are coded as 1

```python
#min total vaccinations for uk
import pandas as pd
url="https://github.com/lilaceri/Working-with-data-/blob/342abab10d93c4bf23b5c55a50f189f12a137c5f/Data%20Sets%20for%20code%20divisio/Covid%20Vaccination%20Data.xlsx?raw=true"
covid=pd.read_excel(url,sheet_name="by_country")
covid.head()
covid1=covid[['total_vaccinations','country']].groupby('country').min() #minimum total_vaccinations for each country

min=covid1.loc["United Kingdom"]['total_vaccinations'] #minimum total_vaccinations for UK

min=int(min) #round value for min total uk vaccinations
min

#function to normalise total_vaccinations column
def normalise_vaccinations(df):
  if df['total_vaccinations']<min:
    return 0
  else:
    return 1

covid.dropna(subset=['total_vaccinations'])
covid['total_vaccinations']=covid.apply(normalise_vaccinations,axis=1)
covid['total_vaccinations']
```

```python
#Display the countries for which total vaccinated is at the same rate or more than the UK
import pandas as pd
url="https://github.com/lilaceri/Working-with-data-/blob/342abab10d93c4bf23b5c55a50f189f12a137c5f/Data%20Sets%20for%20code%20divisio/Covid%20Vaccination%20Data.xlsx?raw=true"
covid=pd.read_excel(url,sheet_name="by_country")
covid1=covid[['total_vaccinations','country']].groupby('country').sum() #sum of total_vaccinations for each country

total=covid1.loc["United Kingdom"]['total_vaccinations'] #find out total vaccinations for UK
print(total)

#find countries for which total vaccinated is at the same rate or more than the UK.
country=[]
for i in covid1.index:
  if covid1.loc[i]['total_vaccinations']>=total and i !='United Kingdom': 
    country.append(i)

country
```

Below, I have written a function which will take a date and display it in parts as specified. I used datetime function from datetime library.I used .year,.month to display year,month of the given date. Used strftime function to display date in differnt formats.

```python
#display parts of date/time
from datetime import datetime
def display_date_parts(date_to_check):
  print("Given date:",date_to_check)
  print("Year:",date_to_check.year)
  print("Month in number:",date_to_check.month)
  print("Month name:",date_to_check.strftime("%B"))
  print("Week number of the year:",date_to_check.strftime("%U"))
  print("Weekday of the week in number:",date_to_check.strftime("%w"))
  print("Day of the year:",date_to_check.strftime("%j"))
  print("Day of the month:",date_to_check.strftime("%d"))
  print("Day of the week as name:",date_to_check.strftime("%A"))






check_date=datetime(1998,12,23)
display_date_parts(check_date)

  
```

Below, I have written a function that takes a date and generate the dates of the following 6 days. To generate the days, I have used the function datetime.timedelta()

```python
#Print dates of the 6 days followed by given date
from datetime import timedelta,datetime
def generate_nextSix_days(date_passed):
  #ini_time_for_now + timedelta(days = 730)
  for i in range(1,7):
    new_date=date_passed+timedelta(days=i)
    print(new_date)
    print("Day of the week:",new_date.strftime("%w"),",Month:",new_date.strftime("%B"),",Year:",new_date.year)


date=datetime(2020,12,28)
generate_nextSix_days(date)
```

**Data Wrangling**

Below I have written a function which
reads the three sheets into the dataframes county_level_df, state_level_df and income_new_df
cleans up the Population column of county_level_df so that it has the name Pop
cleans up the Pop column of state_level_df so that it contains values that are whole numbers of 1000s
combines county_level_df and state_level_df
appends income_new_df to the combined dataframe
SORTs the final dataframe on Income in descending order and
return the final dataframe

```python
import pandas as pd

def get_income_df(url):
  county_level_df=pd.read_excel(url,sheet_name="county-level")

  state_level_df=pd.read_excel(url,sheet_name="state-level")

  income_new_df=pd.read_excel(url,sheet_name="income")

  county_level_df['Pop']=county_level_df['Population']  #to change name of column
  county_level_df.drop(['Population'],axis=1)

  state_level_df['Pop']=round(state_level_df['Pop']/1000) #to convert values to whole number
  state_level_df['Pop']=state_level_df['Pop'].astype(int)

  combined_df=pd.concat([county_level_df,state_level_df], join='inner', ignore_index=True) #to combine two dataframes
   
  income_df_v=income_new_df.append(combined_df,ignore_index=True) #to add new dataframe to combined dataframe
 

  income_df_v4=income_df_v.sort_values(['Income'],ascending=False) #to sort dataframe by Income in descending order
  
  return income_df_v4



url="https://github.com/futureCodersSE/working-with-data/blob/main/Data%20sets/Income-Data.xlsx?raw=true"
income_df=get_income_df(url)
income_df

```

Below I have run the function get_income_df() and store the result in a new dataframe called income_df.ANd then grouped income_df by State and calculated some statistics on the groups.

```python
#to get avearge of income and population for each state

income_df=get_income_df(url)
income_df[['State','Income','Pop']].groupby(['State']).mean()
```

```python
#to get lowest of income and population for each state

income_df=get_income_df(url)
income_df[['State','Income','Pop']].groupby(['State']).min()
```

```python
#to get highest of income and population for each state

income_df=get_income_df(url)
income_df[['State','Income','Pop']].groupby(['State']).max()
```

```python
#to get statistics of Income, Age, Population for each state

income_df=get_income_df(url)
income_df.groupby('State').agg(
    {
       "Age" :['min','max','mean','std','median'],
       "Income":['min','max','mean','std','median'],
       "Pop" :['min','max','mean','std','median']
    }
)
```
